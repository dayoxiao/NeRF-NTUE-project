```python
project_name = 'dataset_20240623_145958'
video_name = '20240623_145958.mp4'
```

```python
import os
import cv2
def extract_images(video_path, output_folder):
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    output_path = output_folder
    if not os.path.exists(output_path):
        os.makedirs(output_path)
    video = cv2.VideoCapture(video_path)
    frame_interval = int(12)
    count = 0
    while video.isOpened():
        ret, frame = video.read()
        if ret:
            if count % frame_interval == 0:
                image_name = os.path.join(output_path, f"{video_name}_{count//frame_interval}.jpg")
                cv2.imwrite(image_name, frame)
            count += 1
        else:
            break
    video.release()
if __name__ == '__main__':
    video_path = f'{video_name}'
    output_folder = f'{project_name}/images'
    extract_images(video_path, output_folder)
            
```
